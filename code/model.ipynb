```python
import numpy as np
import pandas as pd
import networkx as nx
```

```python
ht_result = pd.read_csv("../result_data/ht_result.csv",index_col=0)
hs_result = pd.read_csv("../result_data/hs_result.csv",index_col=0)
```

```python
ht_result[['h', 't']] = ht_result["ht"].str.split(",", expand=True)
new_ord = ['h', 't'] + [col for col in ht_result.columns if col not in ['h', 't', 'ht']]
ht_result = ht_result[new_ord]

hs_result[['h', 's']] = hs_result["hs"].str.split(",", expand=True)
new_ord = ['h', 's'] + [col for col in hs_result.columns if col not in ['h', 's', 'hs']]
hs_result = hs_result[new_ord]
```

```python
ht_result.columns = ['h', 't'] + [x+'2' for x in ht_result.columns if x not in ['h', 't']]
```

```python
import pandas as pd

# Sample data
hs_data = {
    'h': [1, 1],
    's': ['A', 'B'],
    'fe1_1': [0.1, 0.2],
    'fe1_2': [0.4, 0.5],
}

ht_data = {
    'h': [1, 1],
    't': ['X', 'Y'],
    'fe2_1': [0.7, 0.8],
    'fe2_2': [1.0, 1.1],
}

# Convert data to pandas DataFrames
hs_df = pd.DataFrame(hs_data)
ht_df = pd.DataFrame(ht_data)

# Merge datasets on 'h'
merged_df = pd.merge(hs_df, ht_df, on='h')

# Create the final dataset with the required columns
final_df = merged_df[['h'] + [f'fe1_{i}' for i in range(1, 3)] + [f'fe2_{i}' for i in range(1, 3)] + ['t']]

# Display the final dataset
print(final_df)
```

```python
# Merge datasets on 'h'
merged_df = pd.merge(hs_result, ht_result, on='h')

# Create the final dataset with the required columns
final_df = merged_df[['h'] + [col for col in hs_result.columns if col not in ['h', 's']]\
                    + [col for col in ht_result.columns if col not in ['h', 't']] + ['t']]

# Display the final dataset
final_df
```

```python
from sklearn.preprocessing import LabelEncoder

h_labelEncode = LabelEncoder().fit(final_df["h"])
t_labelEncode = LabelEncoder().fit(final_df["t"])

final_df['h'] = h_labelEncode.transform(final_df["h"])
final_df['t'] = t_labelEncode.transform(final_df["t"])
```

```python
final_df
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=0.99)
df_pca = pca.fit_transform(final_df.drop(["h", "t"], axis=1))
```

```python
df_pca.shape
```

```python
X = pd.concat((final_df['h'], pd.DataFrame(df_pca)), axis=1).to_numpy()
y = final_df['t']
```

```python
from sklearn.model_selection import train_test_split

# split into train and test data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42)
```

```python
from sklearn.svm import LinearSVC

# train the model
model = LinearSVC()
model.fit(X_train, y_train)

# predict
y_hat = model.predict(X_test)
```

```python
from sklearn.metrics import f1_score

print(f1_score(y_test, y_hat))
```
